```python
import pandas as pd
import matplotlib.pyplot as plt

# Paths to train and test datasets
train_csv_path = "lyrics_train_set.csv"  # Replace with your train dataset path
test_csv_path = "lyrics_test_set.csv"  # Replace with your test dataset path
```

```python
def plot_songs_per_artist(csv_path, dataset_name):
    """
    Analyze and plot the number of songs per artist in the dataset.

    Args:
        csv_path (str): Path to the dataset CSV file.
        dataset_name (str): Name of the dataset (e.g., "Train" or "Test").
    """
    # Load the dataset
    df = pd.read_csv(csv_path, header=None)

    # Extract the artist column (Column #1)
    artist_counts = df[0].value_counts().head(50)

    # Plot the number of songs per artist
    plt.figure(figsize=(10, 6))
    artist_counts.plot(kind='bar', color='skyblue', alpha=0.8)
    plt.title(f'Number of Songs per Artist in {dataset_name} Dataset', fontsize=16)
    plt.xlabel('Artist', fontsize=12)
    plt.ylabel('Number of Songs', fontsize=12)
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.tight_layout()
    plt.show()

    return artist_counts
```

```python
# Analyze and plot for train dataset
train_artist_counts = plot_songs_per_artist(train_csv_path, "Train")

# Analyze and plot for test dataset
test_artist_counts = plot_songs_per_artist(test_csv_path, "Test")
```

```python
def plot_word_count_distribution(csv_path, dataset_name):
    """
    Analyze and plot the distribution of word counts in the songs.

    Args:
        csv_path (str): Path to the dataset CSV file.
        dataset_name (str): Name of the dataset (e.g., "Train" or "Test").
    """
    # Load the dataset
    df = pd.read_csv(csv_path, header=None)

    # Extract the lyrics column (Column #2) and calculate word counts
    word_counts = df[2].apply(lambda x: len(str(x).split()))

    # Plot the distribution of word counts
    plt.figure(figsize=(10, 6))
    plt.hist(word_counts, bins=50, color='skyblue', alpha=0.8, edgecolor='black')
    plt.title(f'Distribution of Word Counts in {dataset_name} Dataset', fontsize=16)
    plt.xlabel('Number of Words', fontsize=12)
    plt.ylabel('Number of Songs', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

    return word_counts
```

```python
# Plot for train dataset
train_word_counts = plot_word_count_distribution(train_csv_path, "Train")

# Plot for test dataset
test_word_counts = plot_word_count_distribution(test_csv_path, "Test")
```
